```python
#import library 
import quandl
import pandas as pd
import pickle
import matplotlib.pyplot as plt 
import glob 
import numpy as np
import sys
import plotly.plotly as py
import plotly.graph_objs as go
import plotly.plotly as py
py.sign_in('mike2016001', '5ohve2d301')
from mpl_toolkits.basemap import Basemap
from datetime import datetime
start=datetime.now()
from mpl_toolkits.basemap import Basemap
from geopy.geocoders import Nominatim
import math
from geopy.exc import GeocoderTimedOut
```

```python
# read in all the crime data
allFiles1 = glob.glob("*.xls")
allFiles2 = glob.glob("*.xlsx")
allFiles = allFiles1+allFiles2

frame = pd.DataFrame()
lst = []
for file_ in allFiles:
    df = pd.read_excel(file_, index_col = None, header = 0)
    lst.append(df)
frame = pd.concat(lst)
```

```python
#group all the data by state 
state_group_data = frame.groupby(by = 'State')
states= sorted(frame[frame['State'].notnull()]['State'].unique())
```

```python
# calculate all the weapon related crimes during 2012 to 2014

weapon12 = state_group_data['WEAPON12'].count()
weapon13 = state_group_data['WEAPON13'].count()
weapon14 = state_group_data['WEAPON14'].count()

weapon_all = weapon12 + weapon13 + weapon14
```

```python
# plot heatmap by institute name

data = [
    go.Heatmap(
    z = [weapon12,weapon13,weapon14],
    x = states,
    y = ['2012','2013','2014'],
    )

    
    
```
